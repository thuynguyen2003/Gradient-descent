```python
import matplotlib.pyplot as plt
import pandas as pd
plt.subplots()
```

```python
fix, ax = plt.subplots()
plt.show()
```

Câu lệnh plt.subplots(), khi được gọi và không có bất kỳ đầu vào nào, sẽ tạo ra hai đối tượng khác nhau, đối tượng Figure và đối tượng Axes. Đối tượng Figure là một vùng chứa bao gồm các biểu đồ con mà chúng ta thấy. Axes là một biểu đồ con của vùng chứa đó.


```python
records = [('Jan', 42.1), 
           ('Feb', 43.4), 
           ('Mar', 46.6),
           ('Apr', 58.5), 
           ('May', 56.8), 
           ('Jun', 61.8), 
           ('Jul', 65.9), 
           ('Aug', 66.5), 
           ('Sep', 61.6), 
           ('Oct', 53.3), 
           ('Nov', 46.2), 
           ('Dec', 41.1)]
records1 = [('Jan', 4.1), 
           ('Feb', 4.4), 
           ('Mar', 4.6),
           ('Apr', 5.5), 
           ('May', 6.8), 
           ('Jun', 6.8), 
           ('Jul', 5.9), 
           ('Aug', 6.5), 
           ('Sep', 6.6), 
           ('Oct', 5.3), 
           ('Nov', 4.2), 
           ('Dec', 4.1)]

seattle_weather = pd.DataFrame(data = records, columns = ['MONTH', 'MLY-TAVG-NORMAL'])
austin_weather = pd.DataFrame(data = records1, columns = ['MONTH', 'MLY-TAVG-NORMAL'])
seattle_weather
```

Trên đây là cấu trúc dữ liệu DataFrame chứa thông tin về thời tiết ở thành phố Seattle vào các tháng khác nhau trong năm


```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

# Phân chia các biểu đồ con
* Trong một số trường hợp, việc thêm nhiều dữ liệu vào biểu đồ có thể gây rối, khiến chúng ta khó có cái nhìn tổng quát về tập dữ liệu hơn. Do vậy, trong bài học này, chúng ta sẽ cùng tìm cách xử lý vấn đề này.
* Trong các bài trước, chúng ta gọi phương thức subplots() và không truyền bất kỳ đối số đầu vào nào và kết quả trả về là một biểu đồ con. Trong ví dụ bên dưới đây, chúng ta sẽ cung cấp cho hàm một số đối số đầu vào với 3 hàng và 2 cột, ứng với các biểu đồ con.

```python
fig, ax = plt.subplots(3, 2)
plt.show()
```

* Trong trường hợp này, biến ax không còn chỉ là một đối tượng Axes. Thay vào đó, nó là một mảng các đối tượng Axes có kích thước là 3x2. Để thêm dữ liệu, chúng ta sẽ phải sử dụng chỉ số cho các đối tượng này và gọi phương thức plot() trên một phần tử của mảng.

```python
fig, ax = plt.subplots(3, 2)
ax[0][0].plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], color = "r")
ax[0][1].plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"], marker = 'o')
plt.show()
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], color = "b")
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"],
                color='r')
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
plt.show()
```

### Sử dụng phạm vi giá trị
* Để đảm bảo rằng tất cả các biểu đồ con có cùng phạm vi giá trị của trục tung y, chúng ta sẽ khởi tạo vùng chứa biểu đồ và các biểu đồ con bằng đối số sharey có giá trị là True.
* fig, ax = plt.subplots(2, 1, sharey=True)
* Điều này có nghĩa là cả 2 biểu đồ con sẽ có cùng phạm vi giá trị cho trục tung y, dựa trên dữ liệu từ cả 2 tập dữ liệu. Như vậy, việc so sánh giữa các tập dữ liệu sẽ trở nên đơn giản hơn rất nhiều.

```python
fig, ax = plt.subplots(2, 1, sharey = True)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], color = "b")
ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"],
                color='r')
ax[0].set_ylabel("Precipitation (inches)")
ax[1].set_ylabel("Precipitation (inches)")
ax[1].set_xlabel("Time (months)")
plt.show()
```
